Question 8:Standardize the Item Name column. Are there variations that can be unified for better analysis?


```python
df['item_name'] = df['item_name'].str.lower()  # or use str.upper() for uppercase
```

```python
df['item_name'] = df['item_name'].str.strip()
```

```python
df['item_name'] = df['item_name'].str.replace('[^A-Za-z0-9\s]', '_', regex=True)
```

```python
name_mapping = {'cola': 'cola', 'coke': 'cola', 'pepsi': 'cola', 'fizzy drink': 'cola'}
df['Item_Name'] = df['item_name'].replace(name_mapping)
```

```python
df['Item_Name'] = df['Item_Name'].replace({'Ltd.': 'Limited', 'Corp': 'Corporation'}, regex=True)
```

```python
df['Item_Type'] = df['Item_Name'].replace({'Burger': 'Fast Food', 'Pizza': 'Fast Food', 'Salad': 'Healthy'})
```

Question 12:Handling Categorical Data:For categorical columns (e.g., Item Name), consider encoding or transforming them into a format suitable for analysis

```python
one_hot = pd.get_dummies(df, columns=['item_name'], prefix='Item')
```

```python
label = df.copy()
for column in label.columns:
    label['item_name'] = label['item_name'].astype('category').cat.codes
```

```python
print("Original DataFrame:")
print(df)
```
